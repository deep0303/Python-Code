```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
```

```python
start =  datetime.datetime(2014,1,1)
end = datetime.datetime(2024,1,1)
```

```python
BAC = data.DataReader('BAC','stooq',start,end) #bank of America
C = data.DataReader('C','stooq',start,end)
GS = data.DataReader('GS','stooq',start,end)
JPM = data.DataReader('JPM','stooq',start,end)
MS = data.DataReader('MS','stooq',start,end)
WFC = data.DataReader('WFC','stooq',start,end)
```

```python
BAC
```

```python
list_tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=list_tickers)
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# EDA

```python
bank_stocks["BAC"]["Close"].max()#maximum Close price of  Bank of America
```

```python
for tick in list_tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
returns = pd.DataFrame()
```

```python
for tick in list_tickers:
    returns[tick + 'Return'] = bank_stocks[tick]['Close'].pct_change()
print(returns.head())
```

# Moving Averages

```python
returns.sort_index(level=1, inplace=True)
```

```python
plt.figure(figsize=(12, 4))
BAC['30_day_MA'] = BAC['Close'].rolling(window=30).mean()
BAC['30_day_MA'].plot(label='30-day Moving Avg')
BAC['Close'].plot(label='BAC Close')
plt.legend()
plt.title(f'{'BAC'} Closing Price and 30-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
